## Circuits Background
**Circuit with Resistor and LED in Series**

![schematic](img/diag.png)

The circuit we will be working with has the following components:
0. LED - lights up when a high enough current flows through it in the correct direction. Current should flow into the long leg (+) and out of the short leg (-)
0. Resistor - resists the flow of current through it and has a voltage drop across it
0. Power Supply Unit (PSU) - enforces a fixed potential difference between + and - (also called ground) ports
0. Wires - conductive metal which lets current flow through it. Has the same potential along all points on it

Current flows from high to low potential so it'll come out of the + port of the PSU looking for least resistive paths to get to the - port of the PSU. In our schematic, the current will first flow out of the + PSU through the LED (if the LED polarity is correct) then the resistor before reaching the - terminal of the PSU. 


![current flow](img/diag_current.png)

## Exercise 2: Simple Breadboard Use
Now let's try taking advantage of the breadboard to help us make these electrical connections! First make sure the PSU is off.

0. Insert the 4 LED and resistor legs into 4 different rows of the leftside of the breadboard by pushing the legs into holes on the surface of the breadboard.
0. Connect the row with the short LED leg to a row with a resistor leg using a male to male wire
    - Put the ends of the wire into free holes on the 2 rows you wish to connect
0. Clip the red PSU cable to one end of a red male to male wire and do the same with the black PSU wire and a black male to male wire.
0. Insert the red wire into the row with the long LED leg and insert the black wire into the row with the other resistor leg
    - At this point, you should have 2 components inserted into each used row of the breadboard.
0. Turn on the PSU

![ex2_breadboard](img/ex2_breadboard.png)

Did the LED light up again? Why is this also a valid construction of the same circuit schematic? The answer has to do with how a breadboard is internally connected. Each row of 5 holes are electrically connected to each other within the breadboard (underneath the plastic) so we can imagine tiny jumper wires connecting all of the 5 holes to each other. Since these 5 holes are connected, they are all on the same node and should have the same potential.

![breadboard row](img/breadboard_pulled_rail.jpg)

Try:
0. Test the breadboard's internal connections by using the multimeter to find the voltage reading at each free hole on the 4 used rows. Which holes have the same voltage readings? 
0. Move one leg of the LED to a different hole in the same row. Did anything else change?
0. Instead of only using the left half of the breadboard, try moving the LED to the right half and rearrange other wires appropriately to complete the circuit. Is the breadboard internally connected across the gap?
0. Is the male to male jumper wire necessary in this case? Change the position of LED and resistor so you don't need the male to male jumper wire between the LED and resistor. Hint: You should only need 3 rows of the breadboard now

This method seems less likely to fall apart than our first attempt! Using the breadboard makes it easy to build and test circuits quickly and reliably while being less permananent than soldering the connections. 

Take a look at the following schematic:

![5 resistors in parallel](img/5inparallel.png)

What if we now have 5 resistors connected to the - of the PSU? We only have 3 open holes on our ground node/breadboard row so we would need to use a jumper wire to extend the ground node to another row. Putting our PSU + and - nodes into rows of the breadboard limits how many components can access it (without adding in jumpers). It is also hard to access those nodes from any row on the breadboard.


## Exercise 3: Breadboarding with Power Rails
We want to make it the + and - nodes of our circuit more accessible by all components since it is typical for a circuit with many points connected to +/-. In this experiment, we are going to be using the 2 columns of holes (a red +  and blue - column) on the side of your breadboard! These 2 columns are often referred to as the power rails. Instead of horizontal internal connections, the power rails are connected vertically all the way down the side.

![back of breadboard](img/breadboard_back_peel.jpg)


Make sure the PSU is turned off before starting.
0. Move the + and - of the power supply to be in the + and - rail on the left side of the breadboard
0. Use 2 jumper wires to connect the rest of the circuit to the power rails
0. Turn PSU on

![ex 3](img/ex3_breadboard.png)

The LED should have lit up again!

Try:
0. Probe multiple holes in the + and - rail on the breadboard. What are their voltage readings?
0. Were the 2 jumpers that you added necessary? Try moving your resistor and LED legs so that you don't need the jumpers
0. Instead of using the left - rail, try moving one of the jumper wires to use the - rail on the right side of the breadboard. What happened?
    - there are 2 sets of power rails but they are not connected! You may use jumper wires to connect the + rail to the + rail and the - rail to the - rail
